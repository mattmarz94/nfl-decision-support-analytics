```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns

sns.set_theme(style="whitegrid")


def plot_distributions(master_df):

    numeric_cols = [
        "tp_W",
        "tp_PD",
        "off_yds_per_play",
        "off_pts",
        "off_yds",
        "off_turnovers",
        "off_score_pct",
        "off_turnover_pct",
        "off_exp",
        "def_pa",
        "def_yds_allowed",
        "def_yds_per_play",
        "def_turnovers_forced",
        "def_penalty_yds",
        "def_exp"
    ]

    # Only keep columns that actually exist
    numeric_cols = [col for col in numeric_cols if col in master_df.columns]

    for col in numeric_cols:
        plt.figure(figsize=(8,5))
        sns.histplot(master_df[col], bins=20, kde=True)
        plt.title(f"Distribution of {col}")
        plt.xlabel(col)
        plt.ylabel("Count")
        plt.show()

        plt.figure(figsize=(8,4))
        sns.boxplot(x=master_df[col])
        plt.title(f"Boxplot of {col}")
        plt.show()

        print(f"{col} Summary:")
        print(master_df[col].describe())
        print("Skewness:", round(master_df[col].skew(), 3))
        print("-" * 50)
```

```python
master_df = pd.read_csv("Team_Stats_Merge_2018_2025.csv")
plot_distributions(master_df)
```

```python
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt

# Load your dataset
master_df = pd.read_csv("Team_Stats_Merge_2018_2025.csv")

# Select only numeric columns
numeric_df = master_df.select_dtypes(include=["int64", "float64"])

print(numeric_df.columns)
```

```python
corr_matrix = numeric_df.corr()

plt.figure(figsize=(14, 10))
sns.heatmap(
    corr_matrix,
    cmap="coolwarm",
    center=0,
    linewidths=0.5
)

plt.title("Correlation Matrix  Team Season Metrics (20182025)")
plt.show()
```

```python
target_corr = corr_matrix[["tp_W"]].sort_values(by="tp_W", ascending=False)

plt.figure(figsize=(6, 10))
sns.heatmap(
    target_corr,
    annot=True,
    cmap="coolwarm",
    center=0
)

plt.title("Correlation with Regular Season Wins (tp_W)")
plt.show()
```

```python
abs_corr = corr_matrix["tp_W"].abs().sort_values(ascending=False)
abs_corr.head(15)
```

```python
import seaborn as sns
import matplotlib.pyplot as plt

# Columns to exclude
# Trivial/colinear columns     will update to master_df later
exclude_cols = [
    "tp_W",
    "tp_MoV",
    "tp_PF",
    "tp_PA",
    "season",
    "off_g",
    "off_pass_G",
    "off_rush_G",
    "def_g",
    "tp_L",
    "tp_T",
    "tp_W-L%",
    "def_pass_G",
    "off_pts",
    "def_pa",
    "tp_PD",
    "tp_SRS",
    "tp_OSRS",
    "tp_DSRS",

    # ---- Collinear offensive passing cluster (DROP these) ----
    "off_pass_AY/A",
    "off_pass_NY/A",
    "off_pass_EXP",
    # optional extra redundants if you have them:
    "off_pass_Y/A",
    "off_pass_Yds",
    "off_pass_Att",
    "off_pass_Cmp",
    "off_pass_TD",
    "def_pass_PD",
    "def_pass_Att"


]

# Drop them safely if they exist
reduced_df = master_df.drop(columns=[c for c in exclude_cols if c in master_df.columns])

# Keep only numeric columns
numeric_df = reduced_df.select_dtypes(include=["int64", "float64"])

# Compute correlation
corr_matrix = numeric_df.corr()

# Extract correlation with playoffs only
playoff_corr = corr_matrix[["made_playoffs"]].drop("made_playoffs")

# Plot
plt.figure(figsize=(6, 12))
sns.heatmap(
    playoff_corr.sort_values(by="made_playoffs", ascending=False),
    annot=True,
    cmap="coolwarm",
    center=0
)

plt.title("Correlation with Playoff Qualification (Filtered Features)")
plt.show()
```

```python
playoff_corr["abs_corr"] = playoff_corr["made_playoffs"].abs()

top_features = playoff_corr.sort_values("abs_corr", ascending=False).head(8)

print(top_features)
```

```python
def kde_and_box(df, col, title_prefix, xlabel=None):
    """Create 2 plots: KDE by made_playoffs + boxplot by made_playoffs."""
    if col not in df.columns:
        print(f" Skipping '{col}' (column not found). Available columns may use different naming.")
        return

    # KDE plot
    plt.figure(figsize=(8, 5))
    sns.histplot(
        data=df,
        x=col,
        hue="made_playoffs",
        kde=True,
        common_norm=False
    )
    plt.title(f"{title_prefix} (KDE) by Playoff Status")
    plt.xlabel(xlabel if xlabel else col)
    plt.ylabel("Density")
    plt.show()

    # Boxplot
    plt.figure(figsize=(8, 4))
    sns.boxplot(
        data=df,
        x="made_playoffs",
        y=col
    )
    plt.title(f"{title_prefix} (Boxplot) by Playoff Status")
    plt.xlabel("Made Playoffs (0 = No, 1 = Yes)")
    plt.ylabel(xlabel if xlabel else col)
    plt.show()


# 1) off_score_pct
kde_and_box(
    master_df,
    col="off_score_pct",
    title_prefix="Offensive Score Percentage",
    xlabel="Offensive Score %"
)

# 2) off_pass_ANY/A
kde_and_box(
    master_df,
    col="off_pass_ANY/A",
    title_prefix="Passing ANY/A",
    xlabel="Pass ANY/A"
)

# 3) def_turnover_pct
kde_and_box(
    master_df,
    col="def_turnover_pct",
    title_prefix="Defensive Turnover Percentage",
    xlabel="Defensive Turnover %"
)

# 4) off_turnovers
kde_and_box(
    master_df,
    col="off_turnovers",
    title_prefix="Offensive Turnovers",
    xlabel="Offensive Turnovers"
)

# 5) def_yds_per_play
kde_and_box(
    master_df,
    col="def_yds_per_play",
    title_prefix="Defensive Yards Per Play",
    xlabel="Defensive Yards Per Play"
)

# 6) def_pass_ANY/A
kde_and_box(
    master_df,
    col="def_pass_ANY/A",
    title_prefix="Defensive Adjusted Net Yards per Attempt Allowed",
    xlabel="Def Pass ANY/A"
)

# 7) off_pass_TD%
kde_and_box(
    master_df,
    col="off_pass_TD%",
    title_prefix="Offensive Passing Touchdown Rate",
    xlabel="Pass TD %"
)

# 8) def_pass_Int%
kde_and_box(
    master_df,
    col="def_pass_Int%",
    title_prefix="Defensive Interception Rate",
    xlabel="Def Pass INT %"
)
```
