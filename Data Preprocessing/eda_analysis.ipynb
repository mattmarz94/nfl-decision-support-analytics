```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns

sns.set_theme(style="whitegrid")


def plot_distributions(master_df):

    numeric_cols = [
        "tp_W",
        "tp_PD",
        "off_yds_per_play",
        "off_pts",
        "off_yds",
        "off_turnovers",
        "off_score_pct",
        "off_turnover_pct",
        "off_exp",
        "def_pa",
        "def_yds_allowed",
        "def_yds_per_play",
        "def_turnovers_forced",
        "def_penalty_yds",
        "def_exp"
    ]

    # Only keep columns that actually exist
    numeric_cols = [col for col in numeric_cols if col in master_df.columns]

    for col in numeric_cols:
        plt.figure(figsize=(8,5))
        sns.histplot(master_df[col], bins=20, kde=True)
        plt.title(f"Distribution of {col}")
        plt.xlabel(col)
        plt.ylabel("Count")
        plt.show()

        plt.figure(figsize=(8,4))
        sns.boxplot(x=master_df[col])
        plt.title(f"Boxplot of {col}")
        plt.show()

        print(f"{col} Summary:")
        print(master_df[col].describe())
        print("Skewness:", round(master_df[col].skew(), 3))
        print("-" * 50)
```

```python
master_df = pd.read_csv("Team_Stats_Merge_2018_2025.csv")
plot_distributions(master_df)
```
